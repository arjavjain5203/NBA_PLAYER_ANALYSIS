```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('nba_dirty_stats.csv')
dirt_df = pd.DataFrame(data)
clean_df = dirt_df.dropna()
print(clean_df)
```

```python
performance=clean_df.sort_values(by='PPG',ascending=False)
print("Order of top performaers")
print(performance)
```

```python
performance[['FG%', '3P%', 'FT%']] = performance[['FG%', '3P%', 'FT%']].apply(pd.to_numeric, errors='coerce')
performance.plot(x='Player', y=['FG%', '3P%', 'FT%'], kind='bar', figsize=(11,5))
plt.ylabel("percentage")
plt.title("Player performence")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
performance.to_csv('top_performers.csv',index=False)
```
